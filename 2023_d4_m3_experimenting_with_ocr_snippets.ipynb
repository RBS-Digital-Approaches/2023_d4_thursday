### About the Tesseract training we'll see in this notebook
This notebook doesn't work through the code for training Tesseract, but there is a series of notebooks in today's `reference` folder that does, if you're interested in seeing more.

Training Tesseract requires sample data, comprising images of the kind of text you want to recognize and "ground truth" transcriptions of that text. I used a script called [`tesstrain`](https://github.com/tesseract-ocr/tesstrain) that helps to automate the training procedure. `Tesstrain` requires images and transcriptions not of pages of text, but of individual *lines* of text.

Even without training, Tesseract could accurately identify lines of type in the page images of *Sophonisba* and report their coordinates as XML [hOCR](https://en.wikipedia.org/wiki/HOCR):

Rather than transcribing the text in the images, myself, I wrote some Python code to extract lines of text from a slightly-tweaked copy of the TEI-encoded transcription from the ECCO-TCP project and match them to Tesseract's untrained output. I then saved each of those lines of text as a separate text file. I used `tesstrain` to create a new language model based on Tesseract's existing English language model.

As we'll see, this wasn't enough to get the kind of reliability I'd like for an OCR modelI just didn't have enough data to feed it. But it does begin to show how you could fine-tune Tesseract to improve results with a particular font or for a few special characters. (Training Tesseract "from scratch" on a completely different language or character set is certainly possible, but is a larger project.)

````python
#Code cell #12
import numpy as np

display_im = cv2.cvtColor(im, cv2.COLOR_BAYER_GR2BGR)

height = im.shape[0]
width = im.shape[1]

trained_boxes = pytesseract.image_to_boxes(im, output_type=Output.DICT)

n_boxes = len(trained_boxes['char'])
for i in range(n_boxes):
    (text,x1,y1,x2,y2) = (trained_boxes['char'][i],trained_boxes['left'][i],
                          trained_boxes['top'][i],trained_boxes['right'][i],
                          trained_boxes['bottom'][i])
    cv2.rectangle(display_im, (x1,height-y1), (x2,height-y2) , (0,255,0), 2)
cv2_imshow(display_im)
````

##An alternate approach to studying typography
What drew me to this line of inquiry was the possibility of directing OCR to something other than simply the recognition of text. I was interested in re-purposing text recognition technology for the study of typography.

The more time I've spent with this, though, the less sure I am that that's the best approach. It seems to me that this may well be better approached as a straight computer vision questionand Carl's shop has already had some success in this line.

The cells below show some of what might be involved.

### Canny edge detection

````python
#Code cell #13
edges = cv2.Canny(im, 0,255)
cv2_imshow(edges)
````

````python
#Code cell #14
show_rectangles = cv2.cvtColor(im, cv2.COLOR_BAYER_GR2BGR)
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 15))
closed = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, rect_kernel)
contours, hierarchy = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
show_hull = edges.copy()
show_hull = cv2.cvtColor(show_hull, cv2.COLOR_BAYER_GR2BGR)
rectangles = []
for contour in contours:
    hull = cv2.convexHull(contour)
    cv2.drawContours(show_hull, [hull], -1, (0, 0, 255), 2)
    rectangles.append(cv2.boundingRect(hull))
cv2_imshow(show_hull)
````

````python
#Code cell #15
show_rectangles = im.copy()
show_rectangles = cv2.cvtColor(show_rectangles, cv2.COLOR_BAYER_GR2BGR)
for rectangle in rectangles :
  start_point = (rectangle[0], rectangle[1])
  end_point = (rectangle[0] + rectangle[2], rectangle[1] + rectangle[3])
  show_rectangles = cv2.rectangle(show_rectangles, start_point, end_point, (0, 255, 0), 2)
cv2_imshow(show_rectangles)
````
